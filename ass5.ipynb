SECTION 3: Categorization and Binning
10. Create a new column named 'Performance_Level' based on
Performance_Rating:
- Rating < 7 → Low
- Rating between 7 and 8.5 → Medium
- Rating above 8.5 → High

```python
df['Performance_Level']=pd.cut(df['Performance_Rating'],bins=[0,7,8.5,10],labels=["Low",'Medium','High'])
```

```python
print(df['Performance_Level'].head(6))
```

SECTION 4: Grouping & Analysis
11. Find the average Monthly_Salary department-wise.
12. Find the average Performance_Rating gender-wise.
13. Count the number of promoted and not promoted employees.
14. Identify which department performs best on average.

```python
print(df.groupby('Department')['Monthly_Salary'].mean())
```

```python
print(df['Promotion_Status'].value_counts())
```

```python
best_dept = df.groupby('Department')['Performance_Rating'].mean().sort_values(ascending=False)
print(best_dept)
```

```python
print("Best Performing Department:", best_dept.idxmax())
```
